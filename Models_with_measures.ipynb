```python
import pandas as pd
import numpy as np
import os
from autogluon.tabular import TabularPredictor
```

## RM

```python
data_rm = pd.read_csv("transdata_1.csv")
```

```python
label = "y_RM"
```

```python
model_rm = TabularPredictor(label=label)
```

```python
model_rm.fit(data_rm)
```

```python
model_rm.save()
```


To load, use: predictor = TabularPredictor.load("AutogluonModels/ag-20231207_165100")

```python
model_rm.leaderboard()
```

```python
model_rm.fit_summary()
```

## Miary jakości

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
from numpy import mean
# and mean error
```

```python
test_data_rm = pd.read_csv("train__transdata_1.csv")
test_rm_y = test_data_rm['y_RM']
test_rm_X = test_data_rm.drop(['y_RM'], axis=1)
test_rm_X.head()
```

```python
rm_predict = model_rm.predict(test_rm_X)
rm_predict
```

```python
r2_rm = r2_score(test_rm_y, rm_predict)
me_rm = mean(test_rm_y - rm_predict)
mae_rm = mean_absolute_error(test_rm_y, rm_predict)
mape_rm = mean_absolute_percentage_error(test_rm_y, rm_predict)
mse_rm = mean_squared_error(test_rm_y, rm_predict)
```

```python
print('R2 dla zmiennej Rm:', r2_rm )
print('ME dla zmiennej Rm:', me_rm )
print('MAE dla zmiennej Rm:', mae_rm )
print('MAPE dla zmiennej Rm:', mape_rm )
print('MSE dla zmiennej Rm:', mse_rm )
```
